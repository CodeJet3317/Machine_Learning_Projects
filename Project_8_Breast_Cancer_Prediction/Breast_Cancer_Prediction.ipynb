# Breast Cancer Prediction System

```python
# Importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

#### Data Collection and processing

```python
# Loading the csv dataset into pandas dataframe
breast_cancer_data = pd.read_csv('dataset/breast_cancer_data.csv')
```

```python
# Printing the first 5 rows
breast_cancer_data.head()
```

Here diagnosis is a Categorical values and Unnamed:32 is null

```python
# Checking the rows and cols
breast_cancer_data.shape
```

```python
# Checking for null values
breast_cancer_data.isnull().sum()
```

```python
# Removing the unnamed 32 column
breast_cancer_data = breast_cancer_data.drop(columns='Unnamed: 32',axis=1)
```

```python
# Checking if the col is removed or not
breast_cancer_data.head()
```

```python
# Checking for null values
breast_cancer_data.isnull().sum()
```

```python
# Checking the discreted values in diaganosis
breast_cancer_data['diagnosis'].value_counts()
```

Here M -> Malignant <br>
B -> Benign <br>
The data is mainly containing the same amount of values therefore not handling it in this project

```python
# Changing the values of B and M to 1 and 0 using the np command
breast_cancer_data['diagnosis'] = breast_cancer_data['diagnosis'].replace({'B':1,'M':0})
```

```python
breast_cancer_data.head()
```

Finding the statistical values


```python
# Finding the statistical values
breast_cancer_data.describe()
```

```python
breast_cancer_data.info()
```

```python
breast_cancer_data.groupby('diagnosis').mean()
```

```python
# Seprating the data in features and target
X = breast_cancer_data.drop(columns='diagnosis',axis=1)
y = breast_cancer_data['diagnosis']
```

```python
# Printing the data
print(X)
```

```python
print(y)
```

```python
# Normalising the data
scaler = StandardScaler()
```

```python
# Normalising the features
X = scaler.fit_transform(X)
```

```python
# Checking the data is normalised or not
print(X)
```

```python
# Splitting the data set in train and test data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
# Checking the shape
print(X.shape,X_train.shape,X_test.shape)
```

#### Training the model

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train,y_train)
```

#### Evaluating the model

```python
train_data_prediction = classifier.predict(X_train)
train_data_accuracy = accuracy_score(y_train,train_data_prediction)
print(f"Accuracy of training data is : {train_data_accuracy}")
```

```python
test_data_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(y_test,test_data_prediction)
print(f"Accuracy of testing data is : {test_data_accuracy}")
```

Making a predictive system

```python
input_data = (842302,17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189)

input_data_as_array = np.asarray(input_data)
input_data_reshaped = input_data_as_array.reshape(1,-1)
input_data_standardized = scaler.transform(input_data_reshaped)

prediction = classifier.predict(input_data_standardized)
print(prediction)
if prediction[0] == 1:
    print('The person is having Benign')
else:
    print('The person is having Malignant')
```
